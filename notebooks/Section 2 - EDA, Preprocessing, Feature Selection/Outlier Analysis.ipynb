## Outlier Analysis - Numerical Variables

We define a function to conduct outlier analysis which will be used to identify extreme values for the numerical variables. 

```python
#Define numeric features for outlier analysis 
numeric_features = ['total_airflow', 't1_n2o', 't1_nh4', 't1_no3', 't1_do', 
                    't1_airflow', 't1_ss', 't1_temp', 't1_po4', 'inflow']

#Dataframe for numeric features 
df_numeric = df_t1[numeric_features].copy()
```

```python
#Function to complete individual variable analysis for outlier and flatline detection

def inspect_variable(df,
                     col,
                     flatline_threshold=120,
                     hampel_window=7,
                     hampel_sigmas=4,
                     show_hampel=False):
    
    print("="*70)
    print(f"INSPECTING VARIABLE: {col}")
    print("="*70)
    
    series = df[col].dropna()
    
    # Calculate Basic Statistics
    print("\n--- Basic Statistics ---")
    display(series.describe())
    print("Skewness:", series.skew())
    print("Kurtosis:", series.kurtosis())
    print("Distinct values:", series.nunique())
    
    # Linear Boxplot
    plt.figure(figsize=(6,4))
    sns.boxplot(y=series)
    plt.title(f"{col} - Linear Boxplot")
    plt.show()

    # Log Boxplot (if positive)
    if (series > 0).all():
        plt.figure(figsize=(6,4))
        sns.boxplot(y=series)
        plt.yscale("log")
        plt.title(f"{col} - Log Scale Boxplot")
        plt.show()
    else:
        print("\nLog plot skipped (non-positive values present).")
    

    # Top 10 Extremes (Robust)
    median = series.median()
    ranked = series.reindex(
        (series - median).abs().sort_values(ascending=False).index
    )
    
    print("\n--- Top 10 Most Extreme (Distance from Median) ---")
    display(ranked.head(10).reset_index(drop=True))

    # Flatline Detection
    groups = (series != series.shift()).cumsum()
    run_lengths = series.groupby(groups).size()
    run_values = series.groupby(groups).first()
    run_start = series.groupby(groups).apply(lambda x: x.index[0])
    run_end = series.groupby(groups).apply(lambda x: x.index[-1])
    
    flat_df = pd.DataFrame({
        "value": run_values,
        "length": run_lengths,
        "start_time": run_start,
        "end_time": run_end
    })
    
    flat_df = flat_df.sort_values("length", ascending=False)
    
    print("\n--- Top 5 Longest Flatlines ---")
    display(flat_df.head(5))
    
    # Flag long flatlines
    long_flatlines = flat_df[flat_df["length"] >= flatline_threshold]
    
    if len(long_flatlines) > 0:
        print(f"\n⚠ Flatlines ≥ {flatline_threshold} detected:")
        display(long_flatlines)
    else:
        print(f"\nNo flatlines ≥ {flatline_threshold} minutes detected.")

    # Optional Hampel Visualization

    if show_hampel:
        print("\n--- Hampel Visualization ---")
        
        rolling_median = series.rolling(window=hampel_window, center=True).median()
        
        mad = lambda x: np.median(np.abs(x - np.median(x)))
        rolling_mad = series.rolling(window=hampel_window, center=True).apply(mad, raw=True)
        
        threshold = hampel_sigmas * 1.4826 * rolling_mad
        upper = rolling_median + threshold
        lower = rolling_median - threshold
        
        difference = np.abs(series - rolling_median)
        outliers = difference > threshold
        
        plt.figure(figsize=(14,6))
        plt.plot(series.index, series, alpha=0.6, label="Raw")
        plt.plot(rolling_median.index, rolling_median, label="Rolling Median")
        plt.plot(upper.index, upper, linestyle="--", label="Upper Bound")
        plt.plot(lower.index, lower, linestyle="--", label="Lower Bound")
        plt.scatter(series.index[outliers],
                    series[outliers],
                    marker='o',
                    label="Detected Spikes")
        plt.legend()
        plt.title(f"{col} - Hampel Spike Detection")
        plt.show()
        
        print("Spikes detected:", outliers.sum())
    
    print("\nInspection complete.")
```
